## Does location affect house size?

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.displot(data = zillow_encode, x = zillow_encode.beds, multiple = 'stack', hue = 'county')
```

```python
fig = plt.figure();   
ax1 = fig.add_subplot(131)
sns.jointplot(data = zillow_encode, y = 'tax_value', x = 'sqft',  kind='reg', height=5, 
              line_kws={'color': 'red'}, ax = ax1);  
ax2 = fig.add_subplot(132);
ax3 = fig.add_subplot(133);


sns.jointplot(data = zillow_encode, y = 'tax_value', x = 'sqft',  kind='reg', height=5, 
              line_kws={'color': 'red'}, ax = ax2)
sns.jointplot(data = zillow_encode, y = 'tax_value', x = 'sqft',  kind='reg', height=5, 
              line_kws={'color': 'red'}, ax = ax3)

plt.show()
```

```python
sns.lmplot(data = zillow_encode, y = 'tax_value', x = 'sqft', hue = 'county', 
           line_kws={'color': 'red'}, col = 'county')
```

```python
sns.lmplot(data = zillow_encode, y = 'tax_value', x = 'beds', hue = 'county', 
           line_kws={'color': 'red'}, col = 'county')
```

```python
sns.lmplot(data = zillow_encode, y = 'tax_value', x = 'baths', hue = 'county', 
           line_kws={'color': 'red'}, col = 'county')
```

```python
from scipy import stats

corr_sqft, p = stats.pearsonr(zillow_encode.sqft, zillow_encode.tax_value)
print(f'Correlation: {corr_sqft}')
print(f'P: {p}')
```

```python
corr_2, p_2 = stats.pearsonr(zillow_encode.beds, zillow_encode.tax_value)
print(f'Correlation: {corr_2}')
print(f'P: {p_2}')
```

```python
corr_3, p_3 = stats.pearsonr(zillow_encode.baths, zillow_encode.tax_value)
print(f'Correlation: {corr_3}')
print(f'P: {p_3}')
```
